# Most Viewed YouTube Videos

```python
import pandas as pd
import os

pwd = os.getcwd()
youtube_df = pd.read_csv(pwd + '\\List of most-viewed YouTube videos.csv')

youtube_df.head()
```

## Create Columns

### Year Column:

```python
test_df = youtube_df.copy().fillna(0)
test_df['Year'] = test_df['Publication date']
for i in range(len(test_df)):
    test_df['Year'][i] = test_df['Publication date'][i].split('-')[0]





test_df
```

```python
test_df['Views'] = test_df['Views (billions)'] * 1000000000
test_df
```

```python
test_df['Artist'] = test_df['Uploader']

for i in range(len(test_df)):
    test_df['Artist'][i] = test_df['Uploader'][i].split('–')[0]

for i in range(len(test_df)):
    test_df['Artist'][i] = test_df['Artist'][i].split('-')[0]

test_df
```

```python
test_df.to_csv('ModifiedYouTube_df.csv')
```
